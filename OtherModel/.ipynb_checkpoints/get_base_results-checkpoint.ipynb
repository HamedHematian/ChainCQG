```python
import pickle
```

```python
bart_base_loc = "outputs/bart-base/generated_text"
bart_large_loc = "outputs/bart-large/generated_text"
t5_base_loc = "outputs/t5-base/checkpoint-10000/generated_text"
t5_large_loc = "outputs/t5-large/checkpoint-6500/generated_text"
```

```python
gpt2_base_loc = "outputs/gpt2-base/generated_text"
gpt2_medium_loc = "outputs/gpt2-medium/generated_text"
```

```python
def process_gpt(data):
    for one in data:
        for key in one:
            one[key] = one[key][3:-4]
```

```python
def get_pickle(loc):
    with open(loc, "rb") as f:
        data = pickle.load(f)
    return data

bart_base = get_pickle(bart_base_loc)
bart_large = get_pickle(bart_large_loc)
t5_base = get_pickle(t5_base_loc)
t5_large = get_pickle(t5_large_loc)

gpt2_base = get_pickle(gpt2_base_loc)
process_gpt(gpt2_base)
gpt2_medium = get_pickle(gpt2_medium_loc)
process_gpt(gpt2_medium)
```

```python
total = []
for i in range(1000):
    one = {}
    one["context"] = bart_base[i]["context"]
    one["ground truth"] = bart_base[i]["label"]
    one["bart-base"] = bart_base[i]["prediction"]
    one["bart-large"] = bart_large[i]["prediction"]
    one["t5-base"] = t5_base[i]["prediction"]
    one["t5-large"] = t5_large[i]["prediction"]
    one["gpt2-base"] = gpt2_base[i]["prediction"]
    one["gpt2-medium"] = gpt2_medium[i]["prediction"]
    
    total.append(one)
```
